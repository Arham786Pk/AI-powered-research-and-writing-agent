```python
# Warning control
import warnings
warnings.filterwarnings('ignore')
```

```python
from crewai import Agent, Task, Crew
```

```python
import os

openai_api_key = ""

os.environ["OPENAI_API_KEY"] = openai_api_key 
os.environ["OPENAI_API_BASE"] = "https://api.mistral.ai/v1"  
os.environ["OPENAI_MODEL_NAME"] = "mistral-small" 
```

```python
planner = Agent(
    role="Content Planner",
    goal="Plan engaging and factually accurate content on {topic}",
    backstory="You're a research expert creating a detailed outline for an article on {topic}. Your outline should be informative and well-structured.",
    allow_delegation=False,
    verbose=True
)
```

```python
writer = Agent(
    role="Content Writer",
    goal="Write a compelling and well-structured blog post on {topic}",
    backstory="You are a professional writer crafting a detailed, engaging article based on the planner’s outline.",
    allow_delegation=False,
    verbose=True
)
```

```python
editor = Agent(
    role="Editor",
    goal="Review and refine the article to align with professional writing standards.",
    backstory="You are an experienced editor ensuring the article follows proper grammar, structure, and journalistic standards.",
    allow_delegation=False,
    verbose=True
)
```

```python
plan = Task(
    description=(
        "1. Research and create an outline for an article on {topic}.\n"
        "2. Identify the target audience and key points.\n"
        "3. Suggest SEO keywords and sources."
    ),
    expected_output="A structured content plan with an outline, key topics, and SEO keywords.",
    agent=planner
)
```

```python
write = Task(
    description=(
        "1. Write a well-structured blog post based on the content plan.\n"
        "2. Use engaging subtitles, include SEO keywords, and maintain flow.\n"
        "3. Ensure each section has 2-3 paragraphs.\n"
    ),
    expected_output="A high-quality, well-written blog post in markdown format.",
    agent=writer
)
```

```python
edit = Task(
    description="Proofread and refine the blog post for grammar, clarity, and readability.",
    expected_output="A polished, error-free article ready for publication.",
    agent=editor
)
```

```python
crew = Crew(
    agents=[planner, writer, editor],
    tasks=[plan, write, edit],
    verbose=2  # To see detailed logs
)
```

```python
topic = "Artificial Intelligence and its Future"
result = crew.kickoff(inputs={"topic": topic})

# Display the result in markdown format
from IPython.display import Markdown
Markdown(result)
```
